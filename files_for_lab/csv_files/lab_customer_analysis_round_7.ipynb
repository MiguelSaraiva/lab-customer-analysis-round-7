```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
from sklearn.metrics import r2_score
from sklearn.preprocessing import MinMaxScaler
from scipy import stats
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
data = pd.read_csv('marketing_customer_analysis.csv')
data
```

```python
categorical = data.select_dtypes('object')
categorical = categorical.drop(['Customer'], axis = 1)
categorical = categorical.drop(['Effective To Date'], axis = 1)
categorical.columns
```

```python
data
```

```python
one_hot_categoricals=[]
for i in range(len(categorical.columns)):
    col_name = categorical.columns[i]
    one_hot_df = pd.DataFrame(OneHotEncoder().fit_transform(data[[col_name]]).toarray())
    new_col_names = [col_name + '___' + str(col_val) for col_val in one_hot_df.columns]
    one_hot_df.columns = new_col_names
    one_hot_categoricals.append(one_hot_df)
```

```python
one_hot_categoricals[3]
```

```python
label_categoricals=[]
for i in range(len(categorical.columns)):
    data[categorical.columns[i]] = LabelEncoder().fit(data[categorical.columns[i]]).transform(data[categorical.columns[i]])
```

```python
categorical
```

```python
numerical = data._get_numeric_data()
numerical.columns = map(str.lower, numerical)



# for i in range(len(numerical.columns)):
#     temp_mean = np.mean(numerical[numerical.columns[i]])
#     numerical[numerical[numerical.columns[i]] <= 0] = temp_mean
#     temp_std = np.std(numerical[numerical.columns[i]])
#     numerical.loc[abs((numerical[numerical.columns[i]] - temp_mean) / temp_std) > 1.5, numerical.columns[i]] = temp_mean
#     numerical.loc[numerical[numerical.columns[i]] <= 0, numerical.columns[i]] = temp_mean
#     arr, lmbda = stats.boxcox(numerical[numerical.columns[i]])
#     numerical[numerical.columns[i]] = arr

#numerical

for i in range(len(numerical.columns)):
    temp_mean = np.mean(numerical[numerical.columns[i]])
    temp_std = np.std(numerical[numerical.columns[i]])
    numerical.loc[abs((numerical[numerical.columns[i]] - temp_mean) / temp_std) > 1.5, numerical.columns[i]] = temp_mean
    numerical.loc[numerical[numerical.columns[i]] <= 0, numerical.columns[i]] = temp_mean
    numerical[numerical.columns[i]] = np.log(numerical[numerical.columns[i]])
```

```python
X = numerical
for i in range(len(label_categoricals)):
    #X = pd.DataFrame(np.concatenate((X, one_hot_categoricals[i]), axis=1))
    X = pd.DataFrame(np.concatenate((X, label_categoricals[i]), axis=1))

X = X.drop(['customer lifetime value'], 1)
X
```

```python
data.head()
```

```python
Y = data['Customer Lifetime Value']
```

```python
X = pd.DataFrame(MinMaxScaler().fit(X).transform(X))
X
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size= 0.2, random_state= 21)
```

```python
model = sm.OLS(y_train, x_train).fit()
```

```python
predictions = model.predict(x_test)
```

```python
r2_score(y_test, predictions)
```

```python
mse = mean_squared_error(y_test, predictions)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```

```python
mae = mean_absolute_error(y_test, predictions)
mae
```
